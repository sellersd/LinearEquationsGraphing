```python
from numpy import random as rd
from matplotlib import pyplot as plt
import numpy as np

def graph(m, b, num):
    x = np.arange(-10, 10, 1)
    plt.xticks(x)

    plt.grid(True)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.plot(x, m*x+b)
    plt.savefig('solution' + str(num) + '.png')

#    plt.show()

def make_tex_eqn(m, b):
    if m != 0:
        latex_eqn = '$y=' + str(m) + 'x + ' + str(b) + '$'
    else:
        latex_eqn = '$y='  + str(b) + '$'
    while latex_eqn in eqns:
        m = rd.randint(-10, 10)
        b = rd.randint(-10,10)
        if m != 0 and b != 0:
            latex_eqn = '$y=' + str(m) + 'x + ' + str(b) + '$'
        elif m == 0 and b != 0:
            latex_eqn = '$y=' + str(b) + '$'
        elif m != 0 and b == 0:
            latex_eqn = '$y=' + str(m) + 'x$'
        else:
            latex_eqn = '$y='  + str(b) + '$'
    return latex_eqn
```

```python
eqns = []
graphs = []

for x in range(10):
    m = rd.randint(-10, 10)
    b = rd.randint(-10, 10)
    eqns.append(make_tex_eqn(m, b))
    graph(m, b, x)
    print(eqns[x])
```
